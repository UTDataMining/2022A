```python
import pandas as pd
df = pd.read_csv('prefectures.csv')   
df[(df['month']== 10) & (df['date']==10)][['prefectureNameJ','testedPositive']].to_csv('patients.csv',  encoding="utf-8",index=False)
```

```python
pref=df["prefectureNameJ"].unique()
```

```python
d={}
for p in pref:
    patients=df[df['都道府県']==p]['患者数（2020年3月28日からは感染者数）'].sum()
    #print(p, patients)
    d[p]=patients
d
```

```python
df=pd.DataFrame(list(d.items()),  columns=['Prefecture', 'Patiants'])
df.to_csv('patiants.csv',  encoding="utf-8",index=False)
```
